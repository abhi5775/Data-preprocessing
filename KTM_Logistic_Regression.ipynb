```python
# encoding the data
```

```python
def LabelEncode(df):
    from sorted_months_weekdays import Month_Sorted_Month as sort_month, Weekday_Sorted_Week as sort_week
    for col in df.columns:
        col_dict = {}
        if df[col].dtype=='object':
            col2=df[col].unique()
            if col=='month':
                col2=sort_month(col2)
            elif col=='week':
                col2=sort_week(col2)
            z=1
            for i in col2:
                col_dict[i]=z
                z+=1
            df[col+"1"]=df[col].map(col_dict)
        df=df.drop(columns=col)
    return df
```

```python
ordinal_var = df[['Phone Type','Current Bike']]
label=LabelEncode(ordinal_var)
df1=label
df1
```

```python
def OneHotEncode(df):
    for col in df.columns:
        if df[col].dtype=='object':
            col2=df[col].unique()
            for i in col2:
                df[i+"_"+col]=np.where(df[col]==i,1,0)
        df=df.drop(columns = col)
    return df
```

```python
nominal_var = df[['Gender','Occupation','Relationship','Response']]
onehott=OneHotEncode(nominal_var)
onehott
```

```python
onehot=onehott.drop(columns=['Female_Gender','Not purchased_Response'])
df2=onehot
df2
```

```python
df3=df1.join(df2,how='inner')
df3
```

```python
## attaching the age column from df to df3
df3['Age']=df['Age']
```

```python
df3=df3
df3
```

```python
# importing required libraries to do Logistic reg, Decision tree and KNN
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# assigning the target variable to y and input variable to x 
y=df3[['Purchased_Response']]
x=df3.drop(columns=['Purchased_Response'])
```

```python
# Scaling of the input data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
scaled_feature=scaler.transform(x)
scaled_feature
```

```python
df_scaled = pd.DataFrame(data = scaled_feature,columns = x.columns)
x=df_scaled
x
```

```python
# splitting data into trained and test data set
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
print(xtrain.head())
print(ytrain.head())
print(xtest.head())
print(ytest.head())
```

```python
# perforing both logistic reg & the Decision tree & KNN
lr=LogisticRegression()
dt=DecisionTreeClassifier()
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
# fitting the trained data for both the models
lr.fit(xtrain,ytrain)
dt.fit(xtrain,ytrain)
knn.fit(xtrain,ytrain)
```

```python
# predicting the y variable by giving test data as input
pred_lr=lr.predict(xtest)
pred_dt=dt.predict(xtest)
pred_knn=knn.predict(xtest)
```

```python
# checking the accuracy of our model
from sklearn.metrics import accuracy_score,precision_score,recall_score,roc_auc_score,f1_score
```

```python
# accuracy of logistic regression
accuracy_score(ytest,pred_lr)
```

```python
# accuracy of Decision tree
accuracy_score(ytest,pred_dt)
```

```python
# accuracy of KNN
accuracy_score(ytest,pred_knn)
```

```python
table=[['Decision Tree',accuracy_score(ytest,pred_dt),precision_score(ytest,pred_dt),recall_score(ytest,pred_dt),roc_auc_score(ytest,pred_dt),f1_score(ytest,pred_dt)]]
table.append(['Logistic Regn',accuracy_score(ytest,pred_lr),precision_score(ytest,pred_lr),recall_score(ytest,pred_lr),roc_auc_score(ytest,pred_lr),f1_score(ytest,pred_lr)])
table.append(['KNN',accuracy_score(ytest,pred_knn),precision_score(ytest,pred_knn),recall_score(ytest,pred_knn),roc_auc_score(ytest,pred_knn),f1_score(ytest,pred_knn)])
```
